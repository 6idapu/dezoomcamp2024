```python
url_list = ['https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-10.csv.gz',
    'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-11.csv.gz',
    'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-12.csv.gz']
```

```python
taxi_dtypes = {
        'VendorID': pd.Int64Dtype(),
        'passenger_count': pd.Int64Dtype(),
        'trip_distance': float,
        'RatecodeID': pd.Int64Dtype(),
        'store_and_fwd_flag': str,
        'PULocationID': pd.Int64Dtype(),
        'DOLocationID': pd.Int64Dtype(),
        'payment_type': pd.Int64Dtype(),
        'fare_amount': float,
        'extra': float,
        'mta_tax': float,
        'tip_amount': float,
        'tolls_amount': float,
        'improvement_sutcharge': float,
        'total_amount': float,
        'congestion_syrcharge': float
    }

parse_dates = ['lpep_pickup_datetime', 'lpep_dropoff_datetime']
```

```python
df=pd.read_csv(url_list[0],sep=',', compression='gzip', parse_dates = parse_dates, dtype=taxi_dtypes)
```

```python
df2=pd.read_csv(url_list[1],sep=',', compression='gzip', parse_dates = parse_dates, dtype=taxi_dtypes)
```

```python
df3=pd.read_csv(url_list[2],sep=',', compression='gzip', parse_dates = parse_dates, dtype=taxi_dtypes)
```
